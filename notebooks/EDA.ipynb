# EDA - US Vehicle Listings

## Introduction
This notebook explores the US vehicle advertisements dataset.
We will look at the data structure, handle missing values and duplicates, and visualize key trends in the data.

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("../vehicles_us.csv")
print("Shape:", df.shape)
df.head()
```

```python
print("Missing values:")
print(df.isnull().sum())
```

### Missing Values
The dataset has missing values in model_year, cylinders, odometer, and paint_color. We fill odometer with the median and remove zero-price listings.

```python
print("Duplicates:", df.duplicated().sum())
df["odometer"] = df["odometer"].fillna(df["odometer"].median())
df = df[df["price"] > 0]
print("Cleaned shape:", df.shape)
```

### Duplicates
No duplicate rows were found. Zero-price listings were removed as they likely represent data entry errors.

```python
fig = px.histogram(df, x="price", nbins=50, title="Distribution of Car Prices")
fig.show()
```

### Price Distribution
Most vehicles are listed between $1,000 and $20,000, with the distribution skewed right. A small number of luxury vehicles are listed at very high prices.

```python
fig = px.histogram(df[df["odometer"] < 300000], x="odometer", nbins=50, title="Odometer Reading Distribution")
fig.show()
```

### Odometer Distribution
Most vehicles have between 50,000 and 150,000 miles. Very few cars exceed 200,000 miles in listings.

```python
fig = px.scatter(df, x="odometer", y="price", color="condition", title="Price vs Odometer by Condition")
fig.show()
```

### Price vs Odometer
There is a clear negative relationship between mileage and price. Cars in better condition command higher prices even at higher mileage.

## Overall Conclusion

In this analysis we explored 50,000+ US vehicle listings. Key findings:

- Prices are mostly between $1,000-$20,000 with a right skew
- Mileage peaks around 100,000-150,000 miles for most listings
- Condition strongly affects price
- Missing data was handled by filling with median or removing bad rows
- Higher mileage generally leads to lower prices

This dataset could be used to build a price prediction model for used vehicles.
