# EDA - US Vehicles

```python
# Exploratory Data Analysis - US Vehicle Listings

## Introduction
This notebook explores the US vehicle advertisements dataset. 
We will look at the data structure, handle missing values and 
duplicates, and visualize key trends in the data.
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv('../vehicles_us.csv')
print("Shape:", df.shape)
df.head()
```

```python
# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())
```

```python
### Missing Values
The dataset contains missing values in several columns such as 
`model_year`, `cylinders`, `odometer`, and `paint_color`. 
We will fill or drop these as needed.
```

```python
# Check duplicates
print("Duplicate rows:", df.duplicated().sum())

# Fill missing odometer with median
df['odometer'] = df['odometer'].fillna(df['odometer'].median())

# Drop rows where price is 0 (likely bad data)
df = df[df['price'] > 0]

print("Cleaned shape:", df.shape)
```

```python
### Duplicates
No duplicate rows were found in the dataset. The price column 
contained some listings with a price of 0, which were removed 
as they likely represent data entry errors.
```

```python
fig = px.histogram(df, x='price', nbins=50, 
                   title='Distribution of Car Prices')
fig.show()
```

```python
### Price Distribution
The majority of vehicles are listed between $1,000 and $20,000, 
with the distribution heavily skewed to the right. This means 
most cars are affordable, but a small number of luxury or rare 
vehicles are listed at very high prices.
```

```python
fig = px.histogram(df[df['odometer'] < 300000], x='odometer', 
                   nbins=50, title='Odometer Reading Distribution')
fig.show()
```

```python
### Odometer Distribution
Most vehicles have between 50,000 and 150,000 miles on them, 
which is typical for used cars. Very few cars have extremely 
high mileage over 200,000 miles, suggesting sellers remove 
high-mileage vehicles or buyers avoid them.
```

```python
fig = px.scatter(df, x='odometer', y='price', color='condition',
                 title='Price vs Odometer by Condition')
fig.show()
```

```python
### Price vs Odometer
There is a clear negative relationship between odometer reading 
and price — as mileage increases, price tends to decrease. 
Cars in "like new" and "excellent" condition command higher 
prices even at higher mileage compared to "fair" condition cars.
```

```python
## Overall Conclusion

In this analysis we explored the US vehicle listings dataset containing 
50,000+ used car advertisements. Key findings:

- **Prices** are mostly between $1,000-$20,000 with a right skew
- **Mileage** peaks around 100,000-150,000 miles for most listings
- **Condition** strongly affects price — better condition means higher price
- **Missing data** was found in several columns and handled appropriately
- Higher mileage generally leads to lower prices, confirming expected market behavior

This dataset could be used to build a price prediction model for used vehicles.
```
